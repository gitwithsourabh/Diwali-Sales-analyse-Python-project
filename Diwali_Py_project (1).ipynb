```python
ax=sb.countplot(x='Gender', data =df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
axz=df.groupby(['Gender'])['Amount'].sum()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum()

sb.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
ax=sb.countplot(x='Age Group',hue='Gender', data =df)
for i in ax.containers:
    ax.bar_label(i)
```

```python
az=df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sb.barplot(x='Age Group', y='Amount' , data =az)
```

```python
STATE
```

```python
st=df.groupby(['State'], as_index=False)['Orders'].sum()
sb.set(font_scale=0.8) # for text font size
mp.xticks(rotation=70) # for text rotation
sb.barplot(x='State', y='Orders', data=st)
```

```python
st=df.groupby(['State'], as_index=False)['Amount'].sum()
sb.set(font_scale=0.8) # for text font size
mp.xticks(rotation=70) # for text rotation
sb.barplot(x='State', y='Amount', data=st ,order=st.sort_values('Amount').State)
```

```python
xm=sb.countplot(data=df,x='Marital_Status')
for i in xm.containers:
    xm.bar_label(i)
```

```python
st=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum()
sb.barplot(x='Marital_Status',y='Amount',hue='Gender', data=st)
```

```python
ax=sb.countplot(x='Occupation',data=df)
mp.xticks(rotation=70)
for i in ax.containers:
    ax.bar_label(i)
```

```python
ax=df.groupby('Occupation', as_index=False)['Amount'].sum()
mp.xticks(rotation=70)
sb.barplot(x='Occupation',y='Amount',data=ax)
```

```python
ax=sb.countplot(x='Product_Category',data=df)
mp.xticks(rotation=70)
for i in ax.containers:
    ax.bar_label(i)
```

```python
ax=df.groupby('Product_Category', as_index=False)['Amount'].sum()
mp.xticks(rotation=70)
sb.barplot(x='Product_Category',y='Amount',data=ax)
```

```python
ax=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
mp.xticks(rotation=70)
sb.barplot(x='Product_ID',y='Orders',data=ax)
```
